Ahora necesitamos calcular la similitud del coseno entre los elementos del data frame df4

```python
from sklearn.metrics.pairwise import cosine_similarity



# Calcula la similitud del coseno entre todos los pares de elementos
similarity = cosine_similarity(df4.values)

# Imprime la matriz de similitud
similarity
```

Una matriz de similitud es una tabla que mide la similitud entre dos o más objetos. En este caso, la matriz de similitud mide la similitud entre los videojuegos del dataset.

La similitud se calcula utilizando una función de similitud. En este caso, la función de similitud utilizada es la similitud coseno. La similitud coseno mide la similitud entre dos vectores en un espacio vectorial. En el caso de los videojuegos, los vectores representan las características de los videojuegos.

Los valores de la matriz de similitud se encuentran entre 0 y 1. Un valor de 1 indica que dos videojuegos son idénticos, mientras que un valor de 0 indica que dos videojuegos son completamente diferentes.

Ejemplos

Videojuego 1

El videojuego 1 tiene un valor de 1 para todos los videojuegos, lo que significa que es idéntico a sí mismo.

Videojuego 2

El videojuego 2 tiene valores muy similares a los del videojuego 1. Esto significa que los dos videojuegos son muy similares.

Videojuego 3

El videojuego 3 tiene valores ligeramente diferentes a los del videojuego 1. Esto significa que los dos videojuegos son algo similares.

Videojuego 4

El videojuego 4 tiene valores muy diferentes a los del videojuego 1. Esto significa que los dos videojuegos son muy diferentes.

Creamos un modelo de recomendación con el cual ingresamos el id del videojuego y este nos devuelve 5 ids de videojuegos similares.

```python
df5=pd.read_csv('/content/drive/MyDrive/Henry Final_1/dfmachine.csv')
```

Probamos nuestro modelo.

```python
# Define una función para recomendar juegos
def recomendacion_juego(codigo: int):
    # Obtener el índice del juego
    indice = df5.loc[df5['código de videojuegos'] == codigo].index[0]

    # Seleccionar los juegos recomendados
    juegos_recomendados = similitud[indice].sort_values(ascending=False).iloc[1:6].index

    # Obtener los títulos de videojuegos
    titulos_recomendados = df5.loc[juegos_recomendados, ['title', 'código de videojuegos']]

    return titulos_recomendados
```
