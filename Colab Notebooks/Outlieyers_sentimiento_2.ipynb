Un outlier es un valor que se aleja mucho de la tendencia general de los datos, lo que puede indicar un error de medici√≥n, una anomal√≠a o una variaci√≥n natural

Un m√©todo com√∫n para encontrar outliers es usar el rango intercuart√≠lico (IQR), que es la diferencia entre el percentil 75 y el percentil 25 de los datos. Los valores que est√°n por encima o por debajo de 1.5 veces el IQR se consideran outliers. Usando la conveniente funci√≥n pandas .quantile(), podemos crear una simple funci√≥n de Python que tome nuestra columna del dataframe y devuelva los outliers

La siguiente funci√≥n devuelve una serie con los mismos √≠ndices que la columna original, pero solo con los valores que son outliers. Los dem√°s valores no son outliers y son reemplazados por valores nulos.

```python
#crear una funci√≥n para encontrar outliers usando IQR.
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    lower_limit=q1-1.5*IQR
    upper_limit=q3+1.5*IQR
    outliers=df[(df<lower_limit)|(df>upper_limit)]
    return outliers
```

```python
outliers_df=pd.DataFrame()
for col in df_final.select_dtypes(include=np.number).columns:
    outliers_df[col]=find_outliers_IQR(df_final[col])
```

El resultado es un dataframe outliers_df que contiene solo los outliers de cada columna num√©rica dl dataframe df_final.

 Puntaje Z, que es la medida de cu√°ntas desviaciones est√°ndar est√° un valor por encima o por debajo de la media. Los valores que tienen un puntaje Z mayor que 3 o menor que -3 se consideran outliers. Usando la funci√≥n scipy.stats.zscore, podemos crear otra funci√≥n de Python que tome nuestra columna del dataframe y devuelva los outliers

```python
def find_outliers_zscore(df):
    z_scores=stats.zscore(df)
    abs_z_scores=np.abs(z_scores)
    outliers=df[abs_z_scores>3]
    return outliers
```

```python
outliers_z=pd.DataFrame()
for col in df_final.select_dtypes(include=np.number).columns:
    outliers_z[col]=find_outliers_zscore(df_final[col])
```

El resultado es un dataframe outliers_z que contiene solo los outliers de cada columna num√©rica , seg√∫n el criterio del puntaje Z.

Ahora realizaremos el an√°lisis de sentimiento para ello importamos la siguiente librer√≠a.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
# Creamos una instancia del analizador de sentimiento
analyzer = SentimentIntensityAnalyzer()
```

```python
# Definimos una funci√≥n que recibe una rese√±a y devuelve un valor num√©rico seg√∫n la escala
def get_sentiment_value(review):
    # Si la rese√±a est√° ausente, devolvemos 1 (neutral)
    if pd.isnull(review):
        return 1
    # Obtenemos los puntajes de sentimiento para la rese√±a
    scores = analyzer.polarity_scores(review)
    # Obtenemos el puntaje compuesto
    compound = scores["compound"]
    # Si el puntaje compuesto es mayor que 0.05, la rese√±a es positiva (2)
    if compound > 0.05:
        return 2
    # Si el puntaje compuesto es menor que -0.05, la rese√±a es negativa (0)
    elif compound < -0.05:
        return 0
    # Si el puntaje compuesto est√° entre -0.05 y 0.05, la rese√±a es neutra (1)
    else:
        return 1
```

```python
# Aplicamos la funci√≥n a la columna reviews del data frame
df_final["sentiment_analysis"] = df_final["review"].apply(get_sentiment_value)

```
