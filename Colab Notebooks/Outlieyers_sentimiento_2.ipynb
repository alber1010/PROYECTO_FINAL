Un outlier es un valor que se aleja mucho de la tendencia general de los datos, lo que puede indicar un error de medición, una anomalía o una variación natural

Un método común para encontrar outliers es usar el rango intercuartílico (IQR), que es la diferencia entre el percentil 75 y el percentil 25 de los datos. Los valores que están por encima o por debajo de 1.5 veces el IQR se consideran outliers. Usando la conveniente función pandas .quantile(), podemos crear una simple función de Python que tome nuestra columna del dataframe y devuelva los outliers

La siguiente función devuelve una serie con los mismos índices que la columna original, pero solo con los valores que son outliers. Los demás valores no son outliers y son reemplazados por valores nulos.

```python
#crear una función para encontrar outliers usando IQR.
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    lower_limit=q1-1.5*IQR
    upper_limit=q3+1.5*IQR
    outliers=df[(df<lower_limit)|(df>upper_limit)]
    return outliers
```

```python
outliers_df=pd.DataFrame()
for col in df_final.select_dtypes(include=np.number).columns:
    outliers_df[col]=find_outliers_IQR(df_final[col])
```

El resultado es un dataframe outliers_df que contiene solo los outliers de cada columna numérica dl dataframe df_final.

```python
outliers_df
```

 Puntaje Z, que es la medida de cuántas desviaciones estándar está un valor por encima o por debajo de la media. Los valores que tienen un puntaje Z mayor que 3 o menor que -3 se consideran outliers. Usando la función scipy.stats.zscore, podemos crear otra función de Python que tome nuestra columna del dataframe y devuelva los outliers

```python
def find_outliers_zscore(df):
    z_scores=stats.zscore(df)
    abs_z_scores=np.abs(z_scores)
    outliers=df[abs_z_scores>3]
    return outliers
```

```python
outliers_z=pd.DataFrame()
for col in df_final.select_dtypes(include=np.number).columns:
    outliers_z[col]=find_outliers_zscore(df_final[col])
```

El resultado es un dataframe outliers_z que contiene solo los outliers de cada columna numérica , según el criterio del puntaje Z.

```python
outliers_z
```

Ahora realizaremos el análisis de sentimiento para ello importamos la siguiente librería.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
# Creamos una instancia del analizador de sentimiento
analyzer = SentimentIntensityAnalyzer()
```

```python
# Definimos una función que recibe una reseña y devuelve un valor numérico según la escala
def get_sentiment_value(review):
    # Si la reseña está ausente, devolvemos 1 (neutral)
    if pd.isnull(review):
        return 1
    # Obtenemos los puntajes de sentimiento para la reseña
    scores = analyzer.polarity_scores(review)
    # Obtenemos el puntaje compuesto
    compound = scores["compound"]
    # Si el puntaje compuesto es mayor que 0.05, la reseña es positiva (2)
    if compound > 0.05:
        return 2
    # Si el puntaje compuesto es menor que -0.05, la reseña es negativa (0)
    elif compound < -0.05:
        return 0
    # Si el puntaje compuesto está entre -0.05 y 0.05, la reseña es neutra (1)
    else:
        return 1
```

```python
# Aplicamos la función a la columna reviews del data frame
df_final["sentiment_analysis"] = df_final["review"].apply(get_sentiment_value)

```
